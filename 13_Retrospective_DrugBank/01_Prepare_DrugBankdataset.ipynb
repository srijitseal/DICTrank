```python
drugbank = pd.read_csv("structure links.csv")
drugbank = drugbank[["Drug Groups", "Name", "SMILES"]].drop_duplicates(subset="SMILES", keep="first").reset_index(drop=True)
drugbank
```

```python
df = drugbank
df
```

```python
# Step 1: Split the 'Drug Groups' column
df['Drug Groups'] = df['Drug Groups'].str.split('; ')

# Step 2: Use one-hot encoding on split values
ohe = df['Drug Groups'].explode().str.get_dummies().groupby(level=0).sum()

# Step 3: If you want to merge the one-hot encoded columns with the original dataframe:
df = df.join(ohe)

df
```
