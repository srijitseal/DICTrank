```python
MOA = pd.read_csv("../data/processed/MOA/MOA_processed.csv.gz", compression = "gzip")
MOA['pert_iname'] = MOA['pert_iname'].str.lower()
MOA
```

```python
interest_MOA = ["myocardial_perfusion_imaging_(MPI)",'cardiac_arrythmia', 'congenital_heart_defects',
           'congestive_heart_failure', 'cardiomyopathy', 'ventricular_tachycardia_(VT)', 'cardiac_arrest',
           'bradycardia', 'endocarditis', 'coronary_heart_disease', 'heart_attack', 'heart_block', "pert_iname", "Standardized_SMILES"]
```

```python
drugbank = pd.read_csv("DrugBank_noDICTrank_compounds.csv")
drugbank
```

```python
drugbank_moa_cols = drugbank.columns.to_list() + interest_MOA 
```

```python
drugbank = pd.merge(drugbank, MOA.drop(columns=["SMILES","Standardized_InChI"]), on="Standardized_SMILES", how='left')
drugbank = drugbank[drugbank_moa_cols]
drugbank
```

```python
drugbank = drugbank.sort_values("pert_iname")
drugbank = drugbank.drop_duplicates(subset=["SMILES"], keep='first')
drugbank
```

```python
drugbank['Name'] = drugbank['Name'].str.lower()
drugbank
```

```python
from rdkit import Chem
from pandarallel import pandarallel
pandarallel.initialize()

# Convert the substructure strings into Mol objects
DICT_Concern_mols = [Chem.MolFromSmiles(smiles) for smiles in DICT_Concern_substructures]

DICT_rank_mols = [Chem.MolFromSmiles(smiles) for smiles in DICT_rank_substructures]

# Remove None values from DICT_Concern_mols
DICT_Concern_mols = [mol for mol in DICT_Concern_mols if mol is not None]

# Remove None values from DICT_rank_mols
DICT_rank_mols = [mol for mol in DICT_rank_mols if mol is not None]


def check_substructure_presence(smiles_str, sub_mol):
    """Check if a specific substructure is present in the given SMILES string."""
    if not isinstance(smiles_str, str):
        return False
    mol = Chem.MolFromSmiles(smiles_str)
    if mol:
        return mol.HasSubstructMatch(sub_mol)
    return False

# Create a new column for each substructure
for idx, sub_mol in enumerate(DICT_Concern_mols):
    col_name = f"Contains_{DICT_Concern_substructures[idx]}"
    drugbank[col_name] = drugbank['SMILES'].parallel_apply(lambda x: check_substructure_presence(x, sub_mol))
    
for idx, sub_mol in enumerate(DICT_rank_mols):
    col_name = f"Contains_{DICT_rank_substructures[idx]}"
    drugbank[col_name] = drugbank['SMILES'].parallel_apply(lambda x: check_substructure_presence(x, sub_mol))

drugbank
```

```python
interest_MOA = ["myocardial_perfusion_imaging_(MPI)", 'cardiac_arrythmia', 'congenital_heart_defects',
                'congestive_heart_failure', 'cardiomyopathy', 'ventricular_tachycardia_(VT)', 'cardiac_arrest',
                'bradycardia', 'endocarditis', 'coronary_heart_disease', 'heart_attack', 'heart_block']
```

```python
from rdkit.Chem import Draw
from IPython.display import display
from PIL import Image

# Columns of interest
interest_MOA = ["myocardial_perfusion_imaging_(MPI)", 'cardiac_arrythmia', 'congenital_heart_defects',
                'congestive_heart_failure', 'cardiomyopathy', 'ventricular_tachycardia_(VT)', 'cardiac_arrest',
                'bradycardia', 'endocarditis', 'coronary_heart_disease', 'heart_attack', 'heart_block']

# Filter out rows based on interest_MOA columns
filtered_df = drugbank[drugbank[interest_MOA].sum(axis=1) >= 1]

# Extract SMILES, compound names, and molecules with the substructure
filtered_data = filtered_df[['SMILES', 'Name']]
mol_data = [(Chem.MolFromSmiles(row['SMILES']), row['Name']) for _, row in filtered_data.iterrows()]

# Adjusted parameters for visualization
dpi = 300
molSize = (300, 300)

# For each molecule, find the substructure, highlight it, and add the name as a legend
img_list = []
for (mol, name), (_, row) in zip(mol_data, filtered_df.iterrows()):
    for sub_mol in DICT_Concern_mols:
        if mol.HasSubstructMatch(sub_mol):
            # Draw the main molecule with highlight
            main_img = Draw.MolToImage(mol, size=molSize, highlightAtoms=mol.GetSubstructMatch(sub_mol), legend=name, dpi=dpi)
            
            # Draw the substructure alone
            sub_img = Draw.MolToImage(sub_mol, dpi=dpi)
            
            # Combine the main molecule and the substructure side-by-side
            combined_width = main_img.width + sub_img.width
            combined_img = Image.new('RGB', (combined_width, main_img.height))
            combined_img.paste(main_img, (0, 0))
            combined_img.paste(sub_img, (main_img.width, (main_img.height - sub_img.height) // 2))
            
            img_list.append((combined_img, row[interest_MOA]))

# Display the combined images along with additional information
for img, moa_data in img_list:
    display(img)
    print(moa_data[moa_data == 1].index.tolist())
```
