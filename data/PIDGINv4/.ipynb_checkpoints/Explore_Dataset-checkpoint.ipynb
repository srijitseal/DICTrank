```python
import pandas as pd
```

```python
import os

directory = "."
datasets = {}

for name in os.listdir(directory):
    
    if name.endswith('.ipynb_checkpoints'):  # Skip Jupyter notebook files
        continue
        
    print(name)     
    
    file_path = os.path.join(name)
    #print(file_path)
    
    if os.path.exists(file_path):  # Check if file exists before reading
        datasets[name[:-4]] = pd.read_csv(file_path, sep='\t')
    else:
        print(f"No matching file found for: {name}")
```

```python
transposed_df = datasets["Cardiotox_sider_target_pred"].T
# Set the column names as a combination of the first and third rows with a '_' in between
transposed_df.columns = transposed_df.iloc[0].astype(str) + '_' + transposed_df.iloc[2].astype(str)
transposed_df = transposed_df.iloc[3:, :]
transposed_df = transposed_df.reset_index()
transposed_df = transposed_df.rename(columns={'index': 'SMILES'})
transposed_df
```

```python
from sklearn.feature_selection import VarianceThreshold
from itertools import compress

def fs_variance(df, threshold:float=0.05):
    """
    Return a list of selected variables based on the threshold.
    """

    # The list of columns in the data frame
    features = list(df.columns)
    
    # Initialize and fit the method
    vt = VarianceThreshold(threshold = threshold)
    _ = vt.fit(df)
    
    # Get which column names which pass the threshold
    feat_select = list(compress(features, vt.get_support()))
    
    return feat_select

selected_Features = fs_variance(transposed_df.iloc[:,1:], threshold=0.005)
print(len(selected_Features))

df = transposed_df[selected_Features]
df
```

```python
df["Standardized_SMILES"] = transposed_df["SMILES"]

# Get a list of the columns
cols = df.columns.tolist()

# Pop the 'SMILES' column and insert it at the second position
cols.insert(0, cols.pop(cols.index('Standardized_SMILES')))

# Reorder the DataFrame columns
df = df[cols]
df
```

```python
df.to_csv("PIDGINv4_processed.csv.gz", index=False, compression="gzip")
```
