```python
import pandas as pd
```

```python
CP = pd.read_csv("CellPainting_overlaps.c.csv.gz", compression="gzip")
CP
```

```python
map_smiles = pd.read_csv("../processed/CellPaintingSMILES/CellPaintingSMILES_processed.csv.gz", compression="gzip")
map_smiles
```

```python
df = pd.merge(CP, map_smiles, left_on="InChICode_standardised", right_on="InChI")
df
```

```python
df = df.drop(columns=['Metadata_mmoles_per_liter', 'index', 'InChI','SMILES', 'InChICode_standardised'])
df
```

```python
# Get a list of the columns
cols = df.columns.tolist()

# Pop the 'SMILES' column and insert it at the second position
cols.insert(1, cols.pop(cols.index('Standardized_InChI')))
cols.insert(0, cols.pop(cols.index('Standardized_SMILES')))

# Reorder the DataFrame columns
df = df[cols]
df
```
