```python
import pandas as pd
```

```python
df = pd.read_csv("Cell_Painting_Median_Features_Median_doses.csv.gz", compression="gzip")
df
```

```python
df = df.rename(columns={"InChICode_standardised": "InChI"})
df
```

```python
from pandarallel import pandarallel
pandarallel.initialize(progress_bar=False)

from rdkit import Chem

def smiles_from_inchi(value):

    try: return Chem.MolToSmiles(Chem.MolFromInchi(value))
    except: return "Cannot_do"
```

```python
df["SMILES"] = df["InChI"].parallel_apply(smiles_from_inchi)

# Get a list of the columns
cols = df.columns.tolist()

# Pop the 'SMILES' column and insert it at the second position
cols.insert(1, cols.pop(cols.index('SMILES')))

# Reorder the DataFrame columns
df = df[cols]
df
```
