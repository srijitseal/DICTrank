```python
import pandas as pd
```

```python
df = pd.read_csv("sider_processed.csv.gz", compression="gzip")
```

```python
df
```

```python
df = df.drop(columns={"smiles"})

# Get a list of the columns
cols = df.columns.tolist()

# Pop the 'SMILES' column and insert it at the second position
cols.insert(0, cols.pop(cols.index('Standardized_SMILES')))
cols.insert(1, cols.pop(cols.index('Standardized_InChI')))

# Reorder the DataFrame columns
df = df[cols]
df
```

```python
df = df.drop_duplicates(subset=["Standardized_SMILES"], keep="first").reset_index(drop=True)
df
```

```python
df.to_csv("../../binarised/sider/sider_binarised.csv.gz", compression="gzip", index=False)
```

```python
df = df[["Standardized_SMILES", "Standardized_InChI", "Cardiac disorders"]]
df
```

```python
df.to_csv("sider_cardiacdisorders_processed.csv.gz", compression="gzip", index=False)
```
