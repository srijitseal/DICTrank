```python
import pandas as pd
```

```python
chemical= pd.read_csv("chemicalinfo.csv", sep="\t")
chemical
```

```python
drug =  pd.read_csv("druginfo.csv", sep="\t")
drug
```

```python
df = pd.merge(chemical, drug)
df = df[["pert_iname", "smiles", "moa", "target", "disease_area", "indication"]].drop_duplicates().reset_index(drop=True)
df
```

```python
# Dictionary to hold new columns
new_cols_data = {}
cols_to_expand = {"moa","target","disease_area","indication"}
for col in cols_to_expand:
    # Split the column by '|' and expand it into separate columns
    expanded_cols = df[col].str.split('|', expand=True)
    
    # For each expanded column, create new columns with binary values
    for idx in range(expanded_cols.shape[1]):
        unique_values = expanded_cols[idx].dropna().unique()
        for val in unique_values:
            # Create a new column name
            new_col_name = val.strip().replace(' ', '_')
            
            # Create a new column in new_cols_data and fill with 0 or 1
            new_cols_data[new_col_name] = df[col].apply(lambda x: 1 if pd.notna(x) and val in x else 0)

# Convert dictionary to dataframe and concatenate with the original dataframe
new_cols_df = pd.DataFrame(new_cols_data)
df = pd.concat([df, new_cols_df], axis=1)

# Drop the original columns
df.drop(columns=cols_to_expand, inplace=True)
```

```python
df = df.rename(columns={"smiles":"SMILES"})
df
```

```python
df.to_csv("../MOA.csv", index=False, compression="gzip")
```
