```python
import pubchempy as pcp
import time

def get_smiles(chemical_name):
    try:
        compounds = pcp.get_compounds(chemical_name, "name")
        if compounds:
            print(compounds[0].isomeric_smiles)
            return compounds[0].isomeric_smiles
        else:
            return "Cannot do"
    except Exception as e:
        print(f"Error fetching data for {chemical_name}: {e}")
        return "Cannot do"

chemical_names = ['acetone', 'benzene', 'ethanol']
smiles_dict = {}

left_out["SMILES"] = left_out["Active Ingredient(s)"].apply(get_smiles)
```
