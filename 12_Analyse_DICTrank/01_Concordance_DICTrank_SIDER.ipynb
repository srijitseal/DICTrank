```python
import pandas as pd
import os

directory = "../data/binarised/"
datasets = {}

for foldername in ["sider_cardiacdisorders", "DICTrank"]: #os.listdir(directory):
    
    if foldername.endswith('.ipynb_checkpoints'):  # Skip Jupyter notebook files
        continue
        
    print(foldername)     
    
    file_path = os.path.join(directory, foldername, foldername + '_binarised.csv.gz')
    #print(file_path)
    
    if os.path.exists(file_path):  # Check if file exists before reading
        datasets[foldername] = pd.read_csv(file_path, compression='gzip')
    else:
        print(f"No matching file found for folder: {foldername}")
```

```python
df1= datasets["sider_cardiacdisorders"]
df1
```

```python
df2=datasets["DICTrank"]
df2
```

```python
# Merge the two datasets on the 'Standardized_SMILES' column
merged_df = df1.merge(df2, on='Standardized_SMILES', suffixes=('_df1', '_df2'))

# Check if the labels from both datasets match for each SMILES
merged_df['match'] = merged_df['DICTrank'] == merged_df['Cardiac disorders']

# Calculate the concordance rate
concordance_rate = merged_df['match'].mean()
```

```python
merged_df
```

```python
print(f"The concordance rate between the two datasets is: {concordance_rate * 100:.2f}%")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Assuming merged_df is the dataframe from previous example

y_true = merged_df['DICTrank']
y_pred = merged_df['Cardiac disorders']

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print(f"The concordance rate between the two datasets is: {concordance_rate * 100:.2f}%")
print(f"Accuracy: {accuracy* 100:.2f}%")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```
