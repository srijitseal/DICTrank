```python
DICTrank = pd.read_csv("../data/binarised/DICTrank/DICTrank_binarised.csv.gz", compression="gzip")
DICTrank
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
import pandas as pd

# Correctly reference the descriptor functions
descriptor_functions = [x[1] for x in Descriptors._descList][:]

# Sample standardized SMILES list
smiles_list = DICTrank.Standardized_SMILES.to_list()

def calculate_descriptors(smiles, descriptor_functions):
    mol = Chem.MolFromSmiles(smiles)
    desc_values = [desc(mol) for desc in descriptor_functions]
    return desc_values

# Calculate descriptors
data = []
for smiles in smiles_list:
    data.append(calculate_descriptors(smiles, descriptor_functions))

# Convert to DataFrame
descriptor_names = [x[0] for x in Descriptors._descList][:]
df = pd.DataFrame(data, columns=descriptor_names, index=smiles_list).reset_index()
df = df.rename(columns={"index": "Standardized_SMILES"})
df
```

```python
# Merge the two datasets on the 'Standardized_SMILES' column
merged_df = DICTrank.merge(df, on='Standardized_SMILES', suffixes=('_df1', '_df2'))
merged_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, mannwhitneyu, kruskal



# Set a colorblind-friendly palette
colors = sns.color_palette("colorblind")
results = []

for label in df.columns[1:]:
    
    # 1. Visualization

    # Set a colorblind-friendly palette
    colors = sns.color_palette("colorblind")
    # Create a bar plot
    #plt.figure(figsize=(4, 4), dpi=100)
    #sns.set(style="white")  # Set the style

    #ax = sns.barplot(x="DICT _ Concern", y=label, data=df, ci=None, palette=colors)

    # Set labels and title
    print(f'Proportion of {label}')
    #ax.set_ylabel(f'Proportion of {label}', fontsize=12)
    #ax.set_xlabel('DICT_class', fontsize=12)
    #ax.set_title(f'Proportion of {label} across DICT_class', fontsize=14)

    # Remove box line, set y-axis on the left
    #ax.spines['top'].set_visible(False)
    #ax.spines['right'].set_visible(False)
    #ax.spines['left'].set_position(('outward', 10))
    #ax.yaxis.set_ticks_position('left')

    #plt.show()

    # 2. Tests
    if merged_df[label].dtype == 'O' or len(merged_df[label].unique()) <= 5:
        # For categorical data
        
        # Create a contingency table
        contingency_table = pd.crosstab(merged_df["DICT _ Concern"], merged_df[label])
        # Perform the chi-squared test
        chi2, p, _, _ = chi2_contingency(contingency_table)
        
        results.append({
            'Label': label,
            'Statistic value': chi2,
            'P value': p,
            'Test': 'Chi-squared'
        })

    # For continuous data
    else:
        # For continuous data
        
        # Perform the Kruskal-Wallis test
        group1 = merged_df[merged_df["DICT _ Concern"] == 'most'][label]
        group2 = merged_df[merged_df["DICT _ Concern"] == 'less'][label]
        group3 = merged_df[merged_df["DICT _ Concern"] == 'no'][label]
        
        h, p = kruskal(group1, group2, group3)
        
        results.append({
            'Label': label,
            'Statistic value': h,
            'P value': p,
            'Test': 'Kruskal-Wallis'
        })

results_df = pd.DataFrame(results)
results_df
```
