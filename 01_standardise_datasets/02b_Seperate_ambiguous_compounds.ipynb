```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/DICTrank/DICTrank_processed.csv.gz", compression="gzip")
```

```python
ambiguous = df[df["DICT _ Concern"]=="ambiguous"].reset_index(drop=True)
ambiguous
```

```python
ambiguous = ambiguous[~ambiguous.Standardized_SMILES.isin(DICTrank_binarised.Standardized_SMILES.to_list())]
```

```python
ambiguous= ambiguous.reset_index(drop=True)
```
