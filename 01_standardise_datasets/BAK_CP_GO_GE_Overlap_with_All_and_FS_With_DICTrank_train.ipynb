```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

def select_features_using_rfe(df, target_column, num_features=10):
    """
    Select top features using RFE with a RandomForestClassifier.
    
    Args:
    - df (pd.DataFrame): DataFrame containing features and target variable
    - target_column (str): Name of the column in df that contains the target variable
    - num_features (int, optional): Number of top features to select. Defaults to 10.
    
    Returns:
    - list of str: List containing names of the selected features
    """
    
    X = df.drop(columns=[target_column])
    y = df[target_column]
    
    estimator = RandomForestClassifier(n_jobs=-1)  # Use all cores
    selector = RFE(estimator, n_features_to_select=num_features, step=1)
    selector = selector.fit(X, y)
    
    return X.columns[selector.support_].tolist()
```
