```python
import pandas as pd

sider_cardiacdisorders = pd.read_csv("../data/binarised/sider_cardiacdisorders/sider_cardiacdisorders_binarised.csv.gz", compression="gzip")
DICTrank = pd.read_csv("../data/binarised/DICTrank/DICTrank_binarised.csv.gz", compression="gzip")
```

```python
sider_cardiacdisorders = sider_cardiacdisorders.rename(columns={"Cardiac disorders": "Cardiotox"})
sider_cardiacdisorders["source"] = "sider_cardiacdisorders"
#sider_cardiacdisorders_inactives = sider_cardiacdisorders[sider_cardiacdisorders["Cardiotox"]==0]
sider_cardiacdisorders_all = sider_cardiacdisorders
```

```python
DICTrank["source"] = "DICTrank"
DICTrank = DICTrank.rename(columns={"DICTrank": "Cardiotox"})
DICTrank = DICTrank [["Standardized_SMILES","Standardized_InChI","Cardiotox","source"]]
DICTrank
```

```python
# Concatenate the dataframes
combined_df = pd.concat([DICTrank, sider_cardiacdisorders_all])

# Sort by 'source' column to give precedence to DICTrank rows
combined_df = combined_df.sort_values(by='source')

# Drop duplicates based on 'Standardized_SMILES', keeping the first occurrence
combined_df = combined_df.drop_duplicates(subset='Standardized_SMILES', keep='first').reset_index(drop=True)

# Sort by 'Cardiotox' column to give precedence to toxic
combined_df = combined_df.sort_values(by='Cardiotox', ascending=False)

# Drop duplicates based on 'Standardized_SMILES', keeping the first occurrence
combined_df = combined_df.drop_duplicates(subset='Standardized_SMILES', keep='first').reset_index(drop=True)

combined_df
```
