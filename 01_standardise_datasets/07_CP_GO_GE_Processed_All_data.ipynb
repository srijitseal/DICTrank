```python
import os
import sys
sys.path.append('/home/ss2686/03_DICTrank')

from scripts.stratified_split_helper import stratified_data_split

import pandas as pd
import pickle

# Loading the dictionary

datasets = {}
splits = {}
dict_rank_test_smiles = []  # New variable to store DICTrank test_SMILES

directory='../data/processed/'
# Load datasets from given directory
for foldername in os.listdir(directory):
        
    if not foldername.startswith('.'):  # Ignore folders starting with a dot
            
        print(foldername)
        file_path = os.path.join(directory, foldername, f"{foldername}_processed.csv.gz")

        if os.path.exists(file_path):
            datasets[foldername] = pd.read_csv(file_path, compression='gzip')
        else:
            print(f"No matching file found for folder: {foldername}")
```

```python
CP_raw = pd.read_csv("../data/CellPainting/Cell_Painting_Median_Features_Median_doses.csv.gz")
CP_raw
```

```python
CP  = CP_raw[CP_raw["InChICode_standardised"].isin(datasets["CellPaintingSMILES"].InChI.to_list())].reset_index(drop=True)
CP
```

```python
CP = pd.merge(CP, datasets["CellPaintingSMILES"], left_on="InChICode_standardised", right_on="InChI")
CP
```

```python
CP = CP.drop(columns=['Metadata_mmoles_per_liter', 'index', 'InChI','SMILES', 'InChICode_standardised'])
CP
```

```python

# Get a list of the columns
cols = CP.columns.tolist()

# Pop the 'SMILES' column and insert it at the second position
cols.insert(0, cols.pop(cols.index('Standardized_SMILES')))
cols.insert(1, cols.pop(cols.index('Standardized_InChI')))

# Reorder the DataFrame columns
CP = CP[cols]
CP
```

```python
CP = CP.groupby("Standardized_SMILES").median().reset_index(drop=False)
CP
```

```python
CP.to_csv("../data/CellPainting/CellPainting_processed.csv.gz", index=False, compression="gzip")
```
