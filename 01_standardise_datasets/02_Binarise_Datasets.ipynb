```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/DICTrank/DICTrank_processed.csv.gz", compression="gzip")
```

```python
ambiguous = df[df["DICT _ Concern"]=="ambiguous"].reset_index(drop=True)
ambiguous
```

```python
df  = df[df["DICT _ Concern"]!="ambiguous"].reset_index(drop=True)
df
```

```python
df["DICT _ Concern"].value_counts()
```

```python
ordered_categories = ["most", "less", "no"]
df["DICT _ Concern"] = pd.Categorical(df["DICT _ Concern"], categories=ordered_categories, ordered=True)
df = df.sort_values(by="DICT _ Concern")
df
```

```python
df.Standardized_SMILES.nunique()
```

```python
df.Standardized_InChI.nunique()
```

```python
df = df.drop_duplicates(subset=["Standardized_SMILES"], keep="first").reset_index(drop=True)
df
```

```python
df["DICT _ Concern"] = df["DICT _ Concern"].astype(str)
```

```python
df.Standardized_SMILES.nunique()
```

```python
df["DICT _ Concern"].value_counts()
```

```python
# Create a mapping dictionary
mapping_dict = {
    "no": 0,
    "less": 1,
    "most": 1
}

# Replace the values in the column based on the mapping
df['DICTrank'] = df['DICT _ Concern'].replace(mapping_dict)
df
```

```python
df["DICT _ Concern"].value_counts()
```

```python
df.dtypes
```

```python
df = df.drop("SMILES", axis=1)
df
```

```python
df.DICTrank.value_counts()
```

```python
df.to_csv("../data/binarised/DICTrank/DICTrank_binarised.csv.gz", compression="gzip", index=False)
```
